* Groupby the origins and summarize to get the station level entrances
* Groupby the destinations and summarize to get the station level entrances
* Merge the origin and destination passengers dataframes together
* Export

```python
origins=ridership_data.groupby(["ID", "START_PLACE_NAME"]).size().reset_index(name="entrance_passenger")
origins
```

```python
destinations=ridership_data.groupby(["ID_1", "END_PLACE_NAME"]).size().reset_index(name="exit_passenger")
destinations
```

```python
combined = pd.concat([origins, destinations], axis=1)
combined
```

```python

combined.to_csv("output/WMATA_entrances_exits_all_riders.csv")
```
